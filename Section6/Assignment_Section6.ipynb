```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
%matplotlib inline
plt.rcParams['figure.figsize']=8,4
warnings.filterwarnings('ignore')
```

```python
mov=pd.read_csv('Section6.csv',encoding='latin1')
```

```python
mov.head()
```

```python
mov.info()
```

```python
mov=mov.rename(columns={'Day of Week':'Weekday','Movie Title':'Movie','Release Date':'ReleaseDate','Adjusted Gross ($mill)':'AdjustedGross'})
```

```python
mov=mov.rename(columns={'Budget ($mill)':'BudgetMillion','Gross ($mill)':'GrossMillion','IMDb Rating':'IMDbRating','MovieLens Rating':'MovieLensRating','Overseas ($mill)':'OverseasMillion'})
```

```python
mov=mov.rename(columns={'US ($mill)':'USMillion','Profit ($mill)':'ProfitMillion','Runtime (min)':'Runtime','Gross % US':'GrossPercentUS','Overseas%':'OverseasPercent','Profit%':'ProfitPercent'})
```

```python
mov2=mov[((mov.Studio=='WB')|(mov.Studio=='Universal')|(mov.Studio=='Sony')|(mov.Studio=='Buena Vista Studios')|(mov.Studio=='Paramount Pictures')|(mov.Studio=='Fox'))&((mov.Genre=='action')|(mov.Genre=='comedy')|(mov.Genre=='adventure')|(mov.Genre=='animation')|(mov.Genre=='drama'))]
```

```python
mov2
```

```python
sns.set(style='darkgrid',palette='mutted',color_codes= True)
```

```python
bxplt= sns.boxplot(data=mov2,x='Genre',y='GrossPercentUS',orient='v',color='lightgrey',showfliers=False)
plt.setp(bxplt.artists,alpha=0.5)
sns.stripplot(x='Genre',y='GrossPercentUS',data=mov2,jitter=True,size=6,linewidth=0,hue='Studio')
bxplt.axes.set_title('Domestic Gross % by Genre',fontsize=30)
bxplt.set_xlabel('Genre',fontsize=20)
bxplt.set_ylabel('Gross % US',fontsize=20)
bxplt.legend(bbox_to_anchor=(1.05,1),loc=2)
```
